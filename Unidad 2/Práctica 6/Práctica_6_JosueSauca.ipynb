```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('covid-05-17-2020.csv')
```

```python
#Se presenta los datos de la base de datos
data
```

```python
#Se verifica si existen datos nulos dentro de la misma
data.isnull().sum()
```

```python
#Se procede a remplaza los valores nulos con NaN
datos = data.replace('NaN', np.nan)
```

```python
#Se presenta la conversion de la tabla
datos
```

```python
#Se llena los datos nulos mediante el metodo ffill, es un metodo que 
#toma la ultima observacion y la propaga
datos = datos.fillna(method='ffill')
```

```python
#Como se ve el dataset ya no esta nulo
datos.isnull().sum()
```

```python
#Se presenta el nuevo dataset con los datos rellenados
datos
```

```python
datos.describe()
```

```python
#Para separar los datos se va a tomar la columna Pronvicia_State, los valores de
#Virginia y California para realiza la correlacion
datosVirginia = datos[(datos.Province_State == 'Virginia')]
datosCalifornia = datos[(datos.Province_State == 'California')]

#Se ve los datos de vignina de la columna Province_State
datosVirginia.head()
```

```python
#Se ve los datos de Californi de la columna Province_State
datosCalifornia.head()
```

```python
"""
Se presenta una grafica respecto a los datos de covid en el estado de California, para lo cual se va a tomar como
primer parametro los datos confirmados de casos, las muertes, luego se va a tomar en cuenta los 
casos activos y se va a graficar
"""
datosObtenidosCalifornia = datosCalifornia[["Confirmed", "Deaths"]].mean(axis=1)
plt.scatter(datosObtenidosCalifornia, datosCalifornia["Active"])
plt.show()
```

```python
"""
Se presenta una grafica respecto a los datos de covid en el estado de Virginia, para lo cual se va a tomar como
primer parametro los datos confirmados de casos, las muertes, luego se va a tomar en cuenta los 
casos activos y se va a graficar
"""
datosObtenidosVirginia = datosVirginia[["Confirmed", "Deaths"]].mean(axis=1)
plt.scatter(datosObtenidosVirginia, datosVirginia["Active"])
plt.show()
```

```python
datos.corr(method='pearson')
```

```python
#Se utiliza la correlacion para los datos de virginia
datosVirginia.corr(method='pearson')
```

```python
#Se utiliza la correlacion para los datos de california
datosCalifornia.corr(method='pearson')
```

```python
plt.scatter(datos['Active'],datos['Province_State'])
plt.xlabel('Estados de USA')
plt.xlabel('Casos activos de covid')
plt.title('Grafica de Casos activos en las estados de USA')
plt.show()
```

```python
import seaborn as sns

californiaDF = datosCalifornia.drop('Province_State', axis=1)
virginiaDF = datosVirginia.drop('Province_State', axis=1)

californiaDF = californiaDF[californiaDF.columns]
virginiaDF = virginiaDF[virginiaDF.columns]
```

```python
wcorr = californiaDF.corr()
sns.heatmap(wcorr)
plt.savefig('attribute_correlation_california.png')
```

```python
im = plt.imread('./attribute_correlation_california.png')
plt.imshow(im)
plt.show()
```

```python
wcorr = virginiaDF.corr()
sns.heatmap(wcorr)
plt.savefig('attribute_correlation_virginia.png')
```

```python
im = plt.imread('./attribute_correlation_virginia.png')
plt.imshow(im)
plt.show()
```
