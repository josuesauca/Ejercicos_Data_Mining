```python
import pandas as pd
```

```python
datos = pd.read_csv("homicidios.csv")
```

```python
datos
```

```python
"""
En este caso vamos a mostrar los datos que son nulos con los valores boolean
que tienen python:

True = Existen valores en la celda
False= No existen valores nulos en la celda
"""
datos.isnull()
```

```python
"""
En este caso vamos a sumar todos los valores nulos,
como se puede observar donde salen 0 significa que 
no existen valores nulos, en los demas si
"""
datos.isnull().sum()
```

```python
"""
En este caso se va a sumar los valores que no son nulos,
como se puede observar sale 3403 datos en los campos,
esto quiere decir que los datos coinciden con el numero
de filas del archivo csv presentado, asi que esta bien
"""
datos.notnull().sum()
```

```python
"""
En este caso se va a eliminar las filas que tienen campos nulos con
el la funcion dropna() 
"""
datos.dropna()
```

```python
"""
Para este caso se va a rellenar los valores nulos con el numero 100 en los 
campos que correspondan
"""
datos.fillna(value=100)
```

```python
"""
Ahora se va a proceder a utilizar el metodo bfill
para los campos nulos que usa el sigueinte valor valido
para rellenarlos
"""
datos.fillna(method="bfill")
```

```python
"""
Ahora se va a proceder a utilizar el metodo ffill
para los campos nulos que usa el valor valido anterior
para llenarlos
"""
datos.fillna(method="ffill")
```

```python
"""
Mostramos la grafica generada con los datos
"""
datos.plot()
```

```python
"""
Guardamos los datos limpiados y mostramos
"""
datos.to_csv('example.csv')
datos.head(5)
```

```python
import seaborn as sns

# Identificamos los missing values visualmente
sns.heatmap(datos.isnull(), cbar=False)
```

```python
# Kdd apartado
from sklearn.impute import KNNImputer
#En este caso el n_neighbors va 10 de acuerdo a los vecinos mas cercaanos
imputer = KNNImputer(n_neighbors=10, weights="uniform")
# Ajustamos el modelo e imputamos los missing values
imputer.fit(datos[["Edad"]])
datos["Edad"] = imputer.transform(datos[["Edad"]]).ravel()

print("Valores perdidos en normalized-losses: " + 
      str(datos['Edad'].isnull().sum()))
```

```python
datos.head()
```

```python
from sklearn.impute import SimpleImputer

simple_imputer = SimpleImputer(strategy='most_frequent')
datos[datos.columns] = simple_imputer.fit_transform(datos[datos.columns])
```
