# Parte 2
1. Realice un balance de los datos tomando como variable clase la columna "Motivo Desaparición", adicionalmente elimine las últimas 4 columnas del data set ya que no representan información importante.

```python
#Eliminar columnas

datos_eliminados = datos_llenos.drop(columns=['Motivo Desaparción Obs.', 'Fecha Desaparición','Situación Actual','Fecha Localización'])
datos_eliminados
```

2. Analice que otras columnas no serian necesarias para realizar un sobremuestreo de clases.

```python
#Se va a eliminar las columnas latitud y altitud del dataset dado, ya que esos son datos referentes
#a un pais, no serian da ayuda en el caso de motivos de desaparecieron

datos_eliminados = datos_eliminados.drop(columns=['Latitud', 'Longitud'])
```

```python
datos_eliminados
```

3. Realice el sobremuestreo de clases utulizando SMOTE.

```python
from imblearn.over_sampling import RandomOverSampler
import numpy as np
import pandas as pd

# Sacamos los datos de la tabla eliminado
X = datos_eliminados.drop('Motivo Desaparción', axis=1)
y = datos_eliminados['Motivo Desaparción']

# Crear el objeto de sobremuestreo
oversampler = RandomOverSampler(random_state=42)

# Aplicar el sobremuestreo
X_resampled, y_resampled = oversampler.fit_resample(X, y)

# Verificar las nuevas distribuciones de clase
print(pd.Series(y_resampled).value_counts())
```

4. Suba nuevo archivo generado con SMOTE al eva y el archivo utlizado para realizar el taller

```python
# Code cell 33
datos_eliminados.to_csv('nuevos_datos_generados.csv')
```
